```python
import pandas as pd

df = pd.read_csv("./caffeine.csv")
df = df.set_index("drink")
print(df)
```

```python
import seaborn as sns
sns.set_theme()
sns.set(rc={"figure.dpi":300}) #이미지 품질 설정
sns.set(rc={"figure.figsize":(6,3)}) ## 크기설정


## Scatter plot 산점도

sns.scatterplot(x="type", y ="Volume (ml)", data=df, hue ='type')
```

```python
sns.histplot( y= "Caffeine (mg)", data = df)
```

```python
sns.histplot( x= "Calories", data = df)
```

```python
sns.histplot( x= "Volume (ml)", data = df, kde = "True")
```

```python
sns.histplot( x= "Calories", data = df, hue = "type")
```

```python
sns.barplot(x="Calories", y="Caffeine (mg)", data =df)
```

```python
sns.boxplot(x="Volume (ml)", y="Caffeine (mg)", data=df)
```
